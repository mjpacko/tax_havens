```python
import pandas as pd
```

```python
df_blacklist = pd.read_csv("../data/clean/blacklist.csv")
df_blacklist.head()
```

```python
df_blacklist = df_blacklist.rename(columns={'Unnamed: 0': 'Country'})
display(df_blacklist)
```

```python
# Identify numerical columns
numerical_columns = df_blacklist.select_dtypes(include=['number']).columns

# Apply the transformation to the headers of numerical columns
df_blacklist.columns = [
    pd.to_datetime(col, format='%d-%m-%Y').strftime('%d-%m-%Y') if col in numerical_columns else col
    for col in df_blacklist.columns
]
```

```python
df_blacklist.dtypes
```

```python
df_blacklist.dtypes
```

```python
# Identify which columns are date-like strings
# We will attempt to convert each column name to a datetime object, and if it works, we know it's a date column
def convert_to_datetime_if_possible(col):
    try:
        # Try converting the column name to a datetime object
        return pd.to_datetime(col, format='%d-%m-%y', errors='raise')
    except (ValueError, TypeError):
        # If conversion fails, return the column name as is
        return col

# Apply this function to each column name
df_blacklist.columns = [convert_to_datetime_if_possible(col) for col in df_blacklist.columns]

df_blacklist.columns
```

```python
df_blacklist.to_csv('blacklist.csv', index=True)
```

```python
df_blacklist = df_blacklist.replace('United Arab Emirites', 'United Arab Emirates')
df_blacklist = df_blacklist.replace('Seychells', 'Seychelles')
```

```python
df_blacklist.to_csv('blacklist_tableau.csv', index=True)
```
