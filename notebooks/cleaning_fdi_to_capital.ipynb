```python
import pandas as pd
```

```python
df  = pd.read_csv("../data/raw/inward_fdi_gross_capital_formation.csv")
df.head()
```

```python
df = df.rename(columns={'Economy_Label': 'Country'})
df.head()
```

```python
df_cleaned = df.dropna(axis=1, how='all')
df_cleaned.head()
```

```python
df_cleaned.columns = ['01-01-'+col[:4] for col in df_cleaned.columns]
```

```python
df_cleaned.head()
```

```python
df_filled = df_cleaned.fillna(0)
df_filled.head()
```

```python
df = df_filled.rename(columns={'01-01-Coun': 'Country'})
display(df)
```

```python
# Identify which columns are date-like strings
# We will attempt to convert each column name to a datetime object, and if it works, we know it's a date column
def convert_to_datetime_if_possible(col):
    try:
        # Try converting the column name to a datetime object
        return pd.to_datetime(col, format='%d-%m-%y', errors='raise')
    except (ValueError, TypeError):
        # If conversion fails, return the column name as is
        return col

# Apply this function to each column name
df.columns = [convert_to_datetime_if_possible(col) for col in df.columns]

df.columns
```

```python
df = df.replace('China, Macao SAR', 'Macau')
```

```python
df = df.replace('Republic of Korea', 'Korea')
df = df.replace('Russian Federation', 'Russia')
df
```

```python
df.to_csv('fdi_to_capital.csv', index=True)
```
